```python
import numpy as np
import matplotlib.pyplot as plt
import random as rd
```

```python
l = 1
x0 = 0.0
n = 10000

X = [] # lista para guardar la posición generada por el siguiente for 
F = []
```

```python
for i in range(n):
    R = rd.random() # para generar números aleatorios
    if 0 < R < 0.5:
        x0 = x0 + l #marchar a la derecha
        #print(x0)
    else:
        x0 = x0 - l # marchar a la izquierda
        #print(x0)
    X.append(x0)
```
