### c)

```python
data = np.loadtxt("manchasolares.txt",float) #Lee los datos del archivo

x = data[:,0] #Toma los datos de la primer columna(mes) para x
y = data[:,1] #Toma los datos de la segunda columna(manchas solares) para y

r = 5

def ykm(y, k, r): #Define la suma de y_(k+m) para un k dado y variando m
    primer_y = max(k - r , 0)  #Define el primer indice m=-r para los y que sumara
    ultimo_y = k + r + 1   #Define el ultimo indice m=+r para los y que sumara
    sum_ykm = sum(y[primer_y:ultimo_y]) #Suma los y_(k+m) entre los dos indices definidos
    return sum_ykm #Devuelve la suma

yk_list = []
for k in range(len(x)): #Ahora va variando los k para todos los datos que tenemos
    Yk = (1/((2*r)+1))*ykm(y, k, r) #Para cada k obtine un Yk, incluyendo ya el producto or el factor definido en la tarea
    yk_list.append(Yk) #Este Yk se guarda en una lista

plt.plot(x, y, linewidth = 1.5,color = 'purple') #Hace la grfica
plt.plot(x, yk_list, linewidth = 1.5,color = 'blue')

plt.xlim(0,1001) #Restringe a los primeros mil datos

plt.title("Manchas solares") #Agrega el ttulo
plt.xlabel("Tiempo(Mes)") #Agrega nombre al eje x
plt.ylabel("Manchas solares") #Agrega nombre al eje y
plt.grid() #Agrega una red por detras de la grfica

plt.legend(["Manchas Solares", "Media mvil"], fontsize=10) #Agrega identificadores para distinguir los datos de las manchas solares y los de la media mvil
plt.show #Muestra la grfica
```
